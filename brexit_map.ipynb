```python
import folium
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('referendum_working.csv')
df
```

```python
#Calculate mean lat and long to center map
mean_lat = df['LAT'].mean()
mean_lon = df['LONG'].mean()
```

```python
map=folium.Map(location=[mean_lat, mean_lon],zoom_start=6, tiles= 'Mapbox bright')
```

```python
def strength_eu_leave_feeling(leave_eu):
    if leave_eu >= 65:
        cat='Strong Leave'
        return cat
    elif 55<= leave_eu <65:
        cat = 'Decisive Leave'
        return cat
    elif 50< leave_eu <55:
        cat = 'Weak Leave'
        return cat
    elif leave_eu <= 35:
        cat='Strong Remain'
        return cat
    elif 35< leave_eu <=45:
        cat = 'Decisive Remain'
        return cat
    elif 45< leave_eu <50:
        cat = 'Weak Remain'
        return cat

def assign_color(strength):
    if strength == 'Strong Leave':
        col = 'darkred'
        return col
    elif strength == 'Decisive Leave':
        col = 'red'
        return col
    elif strength == 'Weak Leave':
        col = 'lightred'
        return col
    elif strength == 'Strong Remain':
        col = 'darkgreen'
        return col
    elif strength == 'Decisive Remain':
        col = 'green'
        return col
    elif strength == 'Weak Remain':
        col = 'lightgreen'
        return col
    
    
```

```python
df['Strength'] = df['Percent Leave'].apply(strength_eu_leave_feeling)
df = df.dropna()
df
```

```python
#Create feature group to add markers to layer control panel
fg=folium.FeatureGroup(name='EU Referendum')
```

```python
#Add voting locations to feature group
for lat,lon,strength, area, percent_leave in zip(df['LAT'], df['LONG'], df['Strength'], df['Area'], df['Percent Leave']):
    fg.add_child(folium.Marker(location=[lat,lon], popup=area + ': ' + str(percent_leave) + '%', icon=folium.Icon(color = assign_color(strength), icon_color = 'white')))
```

```python
#Add fg to map
map.add_child(fg)

#Add layer control panel
map.add_child(folium.LayerControl())


map.save('map_uk.html')
```
