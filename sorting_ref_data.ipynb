```python
import folium
import pandas as pd
import numpy as np
from geopy.geocoders import Nominatim
from geopy.distance import vincenty

df = pd.read_csv('referendum.csv')
df
```

```python
df_reduced = df[['ID', 'Region Code', 'Region', 'Area Code', 'Area', 'Percent Turnout', 'Valid Votes', 'Remain', 'Leave', 'Percent Remain', 'Percent Leave']]
df_reduced
```

```python
df_reduced['Address'] = df_reduced['Area'] + ', ' + 'United Kingdom'
```

```python
df_reduced
```

```python
geolocator = Nominatim()
f = geolocator.geocode('Luton, United Kingdom')
f.latitude
```

```python
def latme(address):
    try:
        l = geolocator.geocode(address, timeout=10)
        l = l.latitude
        return l
    except:
        pass
    
def longme(address):
    try:
        l = geolocator.geocode(address, timeout=10)
        l = l.longitude
        return l
    except:
        pass
```

```python
longme('Trondheim, NO')
```

```python
df_reduced['LAT'] = df_reduced['Address'].apply(latme)
df_reduced
```

```python
df_reduced['LONG'] = df_reduced['Address'].apply(longme)
df_reduced
```

```python
df_reduced.to_csv('referendum_working.csv', sep=',')
```
